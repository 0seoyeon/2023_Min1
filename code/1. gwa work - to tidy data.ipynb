# 데이터 프레임 형태 변경
- '과 하는일'의 데이터 형태가 column 단위로 되어있어 사용하기 어려웁
- 따라서 row 단위로 바꾸고 과에서 담당하고 있는 업무는 하나의 셀에 입력
- 구 이름, 국 이름은 직접 추가해주기

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import os
```

```python
# 데이터 로드
file_dir = '..\\data\\raw data\\과 하는일\\구로구_업무내용(col).xlsx'
df = pd.read_excel(file_dir)
df.head(2)
```

```python
# 우리가 원하는 형태로 바꿔주기
result_df = pd.DataFrame(columns=['과', '하는일'])
result_df

for i in tqdm(range(len(df.columns))):
    gwa = df.columns[i] # 개별 과 접근

    if gwa[-1] == '과':
        content = ', '.join((df.loc[:, gwa].dropna())) # 하는 일을 하나의 문자열로 합치기

        data = [gwa, content] # 추가할 데이터 -> [과, 하는일] 리스트 만들기
        result_df.loc[result_df.shape[0]+1] = data # 마지막 다음 열에 추가

result_df
```

```python
result_df.to_excel("..\\data\\과 하는일 tidy data\\구로구_변경완.xlsx", index=False)
```
