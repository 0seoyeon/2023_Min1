```python
import pandas as pd
import os
```

# 과 하는일 합치기

```python
path = '..\\data\\과 하는일 tidy data'
file_list = os.listdir(path)
len(file_list)
```

```python
# 모든 파일에 대하여 스키마 확인
for i in range(len(file_list)):
    print(file_list[i].split('_')[0], pd.read_excel(path + '\\' + file_list[i]).columns)
```

```python
# 하나의 데이터프레임으로 합치기
result_df = pd.DataFrame()
for i in range(len(file_list)):
    df = pd.read_excel(path + '\\' + file_list[i])
    result_df = pd.concat([result_df, df], axis=0)
result_df.reset_index(drop=True, inplace=True)
```

```python
result_df
```

```python
result_df.to_csv('..\\data\\0. 구_과_하는일_통합.csv', index=False, encoding='utf-8-sig')
```

# 민원 데이터 합치기

```python
path = '..\\data\\raw data\\새올'
file_list = os.listdir(path)
file_list
```

```python
# 최종 데이터프레임
minwon_df = pd.DataFrame()

for i in range(len(file_list)):
    file_dir = path + '\\' + file_list[i] # 하나의 구 csv 파일 경로
    gu_minwon_df = pd.read_csv(file_dir) # 하나의 구 csv 읽기
    print(file_list[i], gu_minwon_df.shape) # 확인차 print 찍기
    minwon_df = pd.concat([minwon_df, gu_minwon_df], axis=0) # minwon_df 아래로 이어붙이기
```

```python
minwon_df.reset_index(drop=True, inplace=True)
```

```python
# 이상한 컬럼이 껴있어서 제거
drop_col = ['title', 'date', 'content']
minwon_df.drop(drop_col, axis=1, inplace=True)
minwon_df
```

```python
minwon_df.to_csv('..\\data\\0. 새올_통합.csv', index=False, encoding='utf-8-sig')
```
